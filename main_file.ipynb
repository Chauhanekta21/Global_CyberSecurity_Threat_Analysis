STEP04: DATA CLEANING AND PREPARATION

In this step, I will:
- Assess numerical features for potential outliers
- Check whether any outliers represent genuine incidents or data quality issues
- Perform any necessary feature engineering if it adds value to the analysis
- Prepare the dataset for exploratory data analysis (EDA)

```python
# A. Created a copy of the original dataset to preserve raw data
# All cleaning and transformations will be done on this copy

df_copy = df.copy()
df_copy
```

```python
# B. Checked for outliers in key numerical features using boxplots

# Function to create a boxplot for a numerical column
def plot_boxplot(col):
    plt.figure(figsize=(8, 5), dpi=150)
    df_copy.boxplot(column=col)
    plt.title(f'Boxplot of {col}')
    plt.ticklabel_format(style='plain', axis='y')
    plt.show()

# Checked for unusually high or low financial loss values
plot_boxplot("Financial Loss (in Million $)")
# Check for unusually high or low numbers of affected users
plot_boxplot("Number of Affected Users")
# Check for unusually high or low incident resolution times
plot_boxplot("Incident Resolution Time (in Hours)")
```

Outlier Assessment Insights:

- No significant outliers were found in the original numerical features.
- This suggests that the values are fairly evenly distributed and do not contain unusually extreme observations.
- This further supports the earlier observation that the dataset may be synthetically generated or carefully curated, as many features show a balanced and uniform distribution.

```python
# C. Created a new feature "Loss Per User" to understand the average financial loss per affected user 
# This helps understand how much loss was caused on average for each affected user  

df_copy['Loss Per User (in $)'] = (
        df_copy['Financial Loss (in Million $)']* 1000000 
        / df_copy['Number of Affected Users']
        ).round(2)

df_copy
```
